## Sulfur Recovery Unit
<br>
<br>
<font size=5>The sulfur recovery unit (SRU) removes environmental pollutants from acid gas streams before they are released into the atmosphere. Furthermore, elemental sulfur is recovered as a valuable by-product.
<br>
<br>
Hydrogen sulfide is particularly dangerous because it prevents the cells of the human body from breathing.
<br>
<br>
<img src="pictures/SRU.jpg"/>
<br>
<br>
- The SRU takes in two kinds of acid gases as input. The first, rich in H2S, called MEA gas, comes from the gas washing plants; the second, called SWS gas, rich in H2S and NH3 (ammonia), comes from the sour water stripping (SWS) plant. 
<br>
<br>
- Acid gases are burnt in reactors, where H2S is transformed into pure sulfur via a partial oxidation reaction with air. Gaseous combustion products from furnaces are cooled, causing the generation of liquid sulfur, which is collected in catch basins, and then passed through high temperature converters, where a further reaction leads to the formation of water vapor and sulfur. 
<br>
<br>
- The remaining, non converted gas (less than 5%), is fed to the Maxisulfur plant for a final conversion phase. The final gas stream (tail gas) from the SRU contains residual H2S and SO2.
<br>
<br>
<img src="pictures/SRU2.jpg"/>
<br>
<br>   
- The main chamber is fed with MEA gas, and combustion is regulated, in air deficiency, by supplying an adequate air flow (AIR_MEA). The secondary combustion chamber is mainly fed with SWS gas and a suitable air flow is provided (AIR_SWS).
<br>
<br>
- The combustion of SWS gas occurs in a separate chamber with excess air, in order to prevent the formation of ammonium salts in the equipment, thereby giving rise to the generation of nitrogen and nitrogen oxides. The gas input flow in the secondary chamber is kept constant by adding some MEA gas (MEA_SPILLING) when the SWS gas flow is too low. Thus, an adequate air flow is added (MEA_SPILLING_AIR). Air flows are controlled by plant operators to guarantee a correct stoichiometric ratio in the tail gas. Control is improved by a closed-loop algorithm which regulates a further air flow (AIR_MEA_2) on the basis of analysis of the tail gas composition.
<br>
<br>
- Air, which supplies oxygen for the reaction, is an important parameter in the conversion of H2S, being responsible for the tail gas composition. In particular, an excessive air flow tends to increase the concentration of SO2 with respect to H2S, whereas a low air flow leads to the opposite situation.
<br>
<br>
- On-line analyzers are used to measure the concentration of both hydrogen sulfide and sulfur dioxide in the tail gas of each sulfur line. The analyzers adopted are able to measure the quantity [H2S]-2[SO2] (where the brackets indicate concentration), in order to monitor the performance of the conversion process and control the air-to-feed ratio in the SRU with the aim of improving the sulfur extraction process.
<br>
<br>
- Hydrogen sulfide and sulfur dioxide frequently cause damage to sensors, which often have to be removed for maintenance. The design of soft sensors able to predict H2S and SO2 concentrations is therefore required, as in the case study

# Z-score
<br>
<font size=5>Simply put, a z-score is the number of standard deviations from the mean a data point is. A common practice  is to discard data that has a Z-score >3 </font>
<br>
<font size=5> It is important to consider that it asumes a normal distribution on the data.</font>
<br>
<br>
<img src="pictures/zscore.jpg"/>
<br>
<br>
<font size=5> In the following cell we will find outliers in yn using z-score.</font>
<br>
<br>

```python
# Finding outliers with z-score in yn
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

#calculate z-score

z = np.abs(stats.zscore(yn))

# threshold of 3 std deviations to conider outlier
threshold = 3

# find outliers and save in a list
anomalies=[]
xanomalies=[]
i=0

for outlier in z:
        if outlier > 3:
            anomalies.append(yn[i])
            xanomalies.append(i)
        i=i+1

# plot data
plt.figure(figsize=(20, 10))
plt.plot(yn)

# Mark outliers
i=0
for j in anomalies:
    plt.plot([xanomalies[i]], [anomalies[i]],'bo')
    i=i+1
plt.show()
```

# Quartiles
<br>
<font size=5>A different criterion is the use on interquartile range. Elements more than 1.5 interquartile ranges above the upper quartile or below the lower quartile. This method is useful when the data is not normally distributed. </font>
<br>
<br>
<img src="pictures/interquartile.jpg"/>
<br>
<br>
<font size=5>Let's find outliers in Y using this method. </font>
<br>
<br>

```python
import seaborn as sns
import numpy as np
plt.figure(figsize=(20, 10))
sns.boxplot(Y)
Q1= np.percentile(Y, 25)  # Q1
Q3= np.percentile(Y, 75)  # Q3
IQR = Q3 - Q1
print(IQR)
anomalies=[]
xanomalies=[]
i=0
for j in Y:
    if (j < (Q1 - 1.5 * IQR)) or j > (Q3 + 1.5 * IQR):
        anomalies.append(Y[i])
        xanomalies.append(i) 
    i=i+1
plt.figure(figsize=(20, 10))
plt.plot(Y)

# Mark outliers
i=0
for j in anomalies:
    plt.plot([xanomalies[i]], [anomalies[i]],'bo')
    i=i+1
plt.show()
```

# Grubbs test
<br>
<font size=5>In statistics, Grubbs's test or the Grubbs test, also known as the maximum normalized residual test or extreme studentized deviate test, is a test used to detect outliers in a univariate data set assumed to come from a normally distributed population.That is, one should first verify that the data can be reasonably approximated by a normal distribution before applying the Grubbs test.
<br>
<br>
Grubbs's test detects one outlier at a time. This outlier is expunged from the dataset and the test is iterated until no outliers are detected. However, multiple iterations change the probabilities of detection, and the test should not be used for sample sizes of six or fewer since it frequently tags most of the points as outliers.
<br>
<br>
The porcedure is the following:
<br>
<br>
Step 1: Order the data points from smallest to largest.
<br>
<br>
Step 2: Find the mean (xÌ) and standard deviation of the data set.
<br>
<br>
Step 3: Calculate the G test statistic using one of the following equations: </font>
<br>
<br>
<font size=5> <center>  $G = \frac{\displaystyle{\min_{1,...,N} } |Y_i-\bar{Y}|}{s}$
<br>
<br> 
<left>Where $\bar{Y}$ is the mean ann $s$ in the standar deviation 
<br>
<br>
A left-tailed test uses the test statistic:
<br>
<br>
<font size=5> <center>  $G= \frac{\bar{Y} - Y_{min} }{s}$
<br>
<br>
<left>Where $Y_{min}$ is the minimum value.
<br>
<br>
For a right-tailed test, use:
<br>
<br>
<font size=5> <center>  $G= \frac{Y_{max} -\bar{Y}}{s}$
<left>
<br>
<br>
<left>Where $Y_{max}$ is the maximum value.
<br>
<br>
 Step 4: If the resulting test statistic G is greater than the critical value, the corresponding value can be regarded to be an outlier.
<br>
